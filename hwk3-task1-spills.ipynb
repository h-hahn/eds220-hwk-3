```python
### ===== FINAL CODE =====
# import pandas
import pandas as pd

# import geopandas
import geopandas as gpd

# import numpy
import numpy as np

# import matplotlib pyplot
import matplotlib.pyplot as plt

# import NY spills data
spills = pd.read_csv('data/Spill_Incidents_20231109.csv', parse_dates = ['Spill Date'])

# import US counties shapefile
counties = gpd.read_file('data/tl_2022_us_county/tl_2022_us_county.shp')

# lowercase column names
spills.columns = spills.columns.str.lower()

# make snake case
spills.columns = spills.columns.str.replace(' ','_')

# select data about petro spills between Jan 1 and Oct 31, 2023
petr_23 = spills[(spills.material_family == 'Petroleum') & (spills.spill_date >= '01/01/2023') & (spills.spill_date <= '10/31/2023')]  

# new dataframe with # of petroleum spills per county
spills_per_county = petr_23.groupby('county').size().reset_index(name = 'n_spills')

# make column names lowercase
counties.columns = counties.columns.str.lower()

# select new york counties
ny_counties = counties[ counties['statefp'] == '36' ]

# select the first item in that st lawrence index tuple using [0]
st_lawrence_index = spills_per_county[spills_per_county.county=='St Lawrence'].index[0]

# update st lawrence name to st. lawrence
spills_per_county.at[st_lawrence_index, 'county'] = 'St. Lawrence'

# change name of name column in counties dataset
ny_counties = ny_counties.rename(columns={'name': 'county'})

# merge spills_per_county and ny_counties
ny_counties = pd.merge(ny_counties,
                       spills_per_county,
                      how = 'inner',
                      on = 'county')

# add 1 spill to Niagara county to make it 89 (located at 31)
ny_counties.at[31, 'n_spills'] = 89

# create a new figure and axis
fig, ax = plt.subplots(figsize = (7,7))

# add new york map
ny_counties.plot(ax = ax,
                column = 'n_spills',
                legend = True,
                cmap = 'plasma',
                legend_kwds={'shrink':0.8, 'label': 'Number of spills'})

# add citation annotation 
ax.annotate("Data: New York State Department of Environmental Conservation, accessed Nov 7, 2023 \nhttps://data.ny.gov/Energy-Environment/Spill-Incidents/u44d-k5fk",
            xy=(0.05, 0.04), # position
            xycoords='figure fraction', 
            fontsize=10)

# add annotation about tuscarora indian reservation
ax.annotate("*One spill within Niagara County occurred in Tuscarora Indian Reservation",
            xy=(0.05, 0.1), # position
            xycoords='figure fraction', 
            fontsize=10)

# add title
ax.set_title('Number of spills in New York in 2023', fontsize = 15)
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

plt.show()
```
